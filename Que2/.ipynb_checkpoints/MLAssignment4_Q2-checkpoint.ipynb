```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from scipy.spatial import distance
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv("in-vehicle-coupon-recommendation.csv")
df
```

```python
#One hot encoding implementation
features=df.columns
features=features[:-1]
target=df.columns[-1]
one_hot_features=pd.get_dummies(df[features])
one_hot_features
```

```python
tsne = TSNE(n_components=2, verbose=2).fit_transform(one_hot_features)
```

```python
plt.figure(figsize=(7,7))
plt.scatter(tsne[:,0], tsne[:, 1], c=df[target])
plt.legend()
plt.show()
```

```python
def splitting(features,target,p):
  splitting_index= int(p * features.shape[0])
  train_X=features[0:splitting_index]
  train_y=target[0:splitting_index]
  test_X=features[splitting_index:]
  test_y= target[splitting_index:]
  return train_X,train_y,test_X,test_y
```

```python
train_X,train_y,test_X,test_y= splitting(one_hot_features,df[target],0.75)
```

```python
train_X
```

```python
test_X
```

```python
def voting_neighbors(label_neighbors):
  count_0=0
  count_1=0
  for i in range(0,len(label_neighbors)):
    if label_neighbors[i][1]==1:
      count_1 += 1
    elif label_neighbors[i][1]==0:
      count_0 += 1
  
  if count_0 >= count_1:
    return 0
  else:
    return 1
```

```python
def loss_function(expected,predicted):
  wrong_predictions = expected != predicted
  sum_of_wrong_predictions = wrong_predictions.sum()
  loss = sum_of_wrong_predictions/expected.shape[0]
  return loss
```

```python
def accuracy(expected,predicted):
  correct_predictions = expected == predicted
  sum_of_correct_predictions = correct_predictions.sum()
  accuracy = sum_of_correct_predictions/expected.shape[0]
  return accuracy
```

Grid Search For K values

```python
loss_dict={}
number_neighbors_list=[10,20,30,40,50,60,70,80,100]
y_predicted=[]
neighbour_dist = distance.cdist(test_X, train_X, 'euclidean')
for x in neighbour_dist:
  neighbor_labels = sorted(list(zip(x,train_y)))
 # print(neighbor_labels)
  different_k_predictions=[]
  for k in number_neighbors_list:
    different_k_predictions.append(voting_neighbors(neighbor_labels[0:k]))
  y_predicted.append(different_k_predictions)
```

```python
y_predicted = np.array(y_predicted)
y_predicted
```

```python
loss_kvise={}
for k in range(0,len(number_neighbors_list),1):
  loss_kvise[number_neighbors_list[k]]= loss_function(test_y,y_predicted[:,k])
```

```python
loss_kvise
```

Running Knn on optimal value of K

Train Accuracy and loss for optimal K(number of nearest neighbours

```python
y_predicted_train=[]
neighbour_dist_train = distance.cdist(train_X, train_X, 'euclidean')
for x in neighbour_dist_train:
  neighbor_labels_train = sorted(list(zip(x,train_y)))
 # print(neighbor_labels)
  y_predicted_train.append(voting_neighbors(neighbor_labels_train[0:20]))
```

```python
loss_function(train_y,y_predicted_train)
```

```python
accuracy(train_y,y_predicted_train)
```

Test Accuracy and loss for optimal k(nearest neighbours)

```python
y_predicted=[]
neighbour_dist = distance.cdist(test_X, train_X, 'euclidean')
for x in neighbour_dist:
  neighbor_labels = sorted(list(zip(x,train_y)))
 # print(neighbor_labels)
  y_predicted.append(voting_neighbors(neighbor_labels[0:20]))
  
```

```python
loss_function(test_y,y_predicted)
```

```python
accuracy(test_y,y_predicted)
```

Error Vs number of neighbours Graph

```python
error_list=[]
for x in number_neighbors_list:
  error_list.append(loss_kvise[x])
```

```python
error_list
```

```python
plt.figure(figsize=(10,10))
plt.plot(number_neighbors_list,error_list)
plt.xlabel('number of neighbours')
plt.ylabel('error')
plt.legend()
plt.plot()
```

Sklearn Knn

```python
kneighbours_sklearn = KNeighborsClassifier(n_neighbors=80)
kneighbours_sklearn.fit(train_X,train_y)
```

```python
kneighbours_sklearn.score(train_X,train_y)
```

```python
kneighbours_sklearn.score(test_X,test_y)
```
